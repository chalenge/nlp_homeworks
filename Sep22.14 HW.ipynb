```python
# -*- coding: utf-8 -*- 
%matplotlib inline

import string
import nltk
import os

from nltk.corpus import brown
sent_tokenizer=nltk.data.load('tokenizers/punkt/english.pickle')
```

```python
currentFile = "../../napoleon_letters.txt"
with open(currentFile, 'r') as book:
    text = book.read()

pattern = r'''(?x)
([A-Z]\.)\s([A-Z]\.)
|([A-Z]\.)+
|\w+([-']\w+)*
|\$?\d+(\.\d+)?%?
|[.,?;]+
'''
words_list = nltk.regexp_tokenize(text,pattern)
words_list
```

```python
brown_news_tagged = brown.tagged_words(simplify_tags=True)
tags = [b[1] for (a, b) in nltk.bigrams(brown_news_tagged) if a[1] == 'V']
fd1 = nltk.FreqDist(tags)
fd1.tabulate()
```

```python
data = nltk.ConditionalFreqDist((word.lower(), tag) for (word, tag) in brown_news_tagged)

for word in data.conditions():
    if len(data[word]) > 3:
        tags = data[word].keys()
        print word, ' '.join(tags)
```
